```python
driver=webdriver.Chrome('C:/Users/kaitnam/Documents/PYTHON/chromedriver.exe')
url=('https://www.baristapaulbassett.co.kr/menu/View.pb?dpid=PB179843')
driver.get(url)
html_source=driver.page_source
driver.close()

resultt=[]
container=dict()
soup=BeautifulSoup(html_source, 'html.parser')
name=str(soup.select('div.menuTit > dl > dt')).split('</span>')[1].split('<')[0]
volume=soup.select_one('span.sizeMl > span').text + 'ml'
span=soup.findAll('span',{"class":"num"})
kcal=span[0].text + 'kcal'
protein=span[1].text + 'g'
sugars=span[2].text + 'g'
sodium=span[3].text + 'mg'
sat_FAT=span[4].text + 'g'
caffeine=span[5].text + 'mg'

container['name']=name
container['volume']=volume
container['kcal']=kcal
container['sugars']=sugars
container['protein']=protein
container['sat_FAT']=sat_FAT
container['sodium']=sodium
container['caffeine']=caffeine
resultt.append(container)
pprint(resultt)
```

```python
import time

result=[]
for number in numbers:
    container=dict()
    driver=webdriver.Chrome('C:/Users/kaitnam/Documents/PYTHON/chromedriver.exe')
    driver.get('https://www.baristapaulbassett.co.kr/menu/View.pb?dpid={}'.format(number))
    html_source=driver.page_source
    soup=BeautifulSoup(html_source, 'html.parser')
    driver.close()
    
    name=str(soup.select('div.menuTit > dl > dt')).split('</span>')[1].split('<')[0]
    volume=soup.select_one('span.sizeMl > span').text + 'ml'
    span=soup.findAll('span',{"class":"num"})
    kcal=span[0].text + 'kcal'
    protein=span[1].text + 'g'
    sugars=span[2].text + 'g'
    sodium=span[3].text + 'mg'
    sat_FAT=span[4].text + 'g'
    caffeine=span[5].text + 'mg'

    container['name']=name
    container['volume']=volume
    container['kcal']=kcal
    container['sugars']=sugars
    container['protein']=protein
    container['sat_FAT']=sat_FAT
    container['sodium']=sodium
    container['caffeine']=caffeine
    result.append(container)
    
    time.sleep(3)

pprint(result)
    
    
```
