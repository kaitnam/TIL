```python
# 분석
# 게스트들은 거의 대부분 1-4일정도를 우리집에서 머물다가 갔다.
```

```python
def parse_month(s):
    return (s[:7])
```

```python
# https://smartreporter3.tistory.com/373
airbnb['month'] = airbnb['숙박일'].apply(parse_month)
airbnb['month']
```

```python
plt.figure(figsize=(20, 5))
sns.countplot(x = 'month', data = airbnb)
```

```python
# 밑에 month가 잘 보이지 않아서 약간 45도 돌려서 나타내려고 한다.

plt.figure(figsize=(20, 5))
p = sns.countplot(x = 'month', data = airbnb)
p.set_xticklabels(p.get_xticklabels(), rotation=45)
```
