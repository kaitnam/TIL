```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib
import pandas as pd

number_list = ['154667', '194334', '187322', '196051', '19335', '202903', '91101', '107992', '37936', '36843', '191920', '17059', '172113']
result=[]
for i in number_list:
    container=dict()
    
    driver=webdriver.Chrome('C:/Users/kaitnam/AppData/Local/Programs/Python/Python38-32/chromedriver.exe')
    driver.get('https://movie.naver.com/movie/bi/mi/basic.nhn?code={}'.format(i))
    html_source=driver.page_source
    soup=BeautifulSoup(html_source, 'html.parser')
    driver.close()


    h_movie = soup.find('h3',{'class':'h_movie'})
    title = h_movie.find('a').text
    point = soup.find("span",{'class':'st_on'}).text.split('객')[1]
    dd_type = soup.find("dl",{'class':'info_spec'})
    d_type = dd_type.findAll('a')[0].text
    country = dd_type.findAll('span')[1].text.replace("\n","").replace("\t","")
    time = dd_type.findAll('span')[2].text
    d_level = dd_type.findAll('dd')[3]
    level = d_level.find('p').text.split(']')[1].replace("\n","").replace("\t","")[:8]
    heartt = soup.find('div',{'class':'u_likeit_module'})
    heart = heartt.find('em').text
    
    container['title']=title
    container['point']=point
    container['movie_type']=d_type
    container['country']=country
    container['time']=time
    container['movie_level']=level
    container['heart']=heart
    result.append(container)


print(result)
```

```python
movie=pd.DataFrame(result)

movie.to_excel("C:/Users/kaitnam/Documents/PYTHON/Project_CoffeeshopMenu/Paulbassett_Menu.xlsx", index=False)
movie
```
