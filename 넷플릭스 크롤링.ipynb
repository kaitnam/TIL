```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib
import pandas as pd

number_list = ['70018292', '81221938', '70270998', '81168885', '70032594', '80990397', '80150243', '1181461', '60024942', '81336395']

result=[]
for k in number_list:
    container=dict()
    driver=webdriver.Chrome('C:/Users/kaitnam/AppData/Local/Programs/Python/Python38-32/chromedriver.exe')
    driver.get("https://www.netflix.com/title/{}".format(k))
    html_source=driver.page_source
    soup = BeautifulSoup(html_source,"html.parser")
    driver.close()

    title = soup.find("h1",{"class": "title-title"}).text
    year = soup.find('span',{'class':"title-info-metadata-item item-year"}).text
    m_num = soup.find ("span",{"class": "maturity-number"}).text
    time = soup.find("span",{"class":"duration"}).text

    dd = soup.findAll("div",{"class":"more-details-item-container"})
    dd = soup.findAll("a",{"class":"more-details-item item-genres"})
    for i in dd:
        m_type = i.text
    rr = soup.findAll('span',{'class':'more-details-item item-mood-tag'}) #for문을 돌렸는데 한개밖에 안나옴.
    for j in rr:
        spec = j.text
    #print(spec)
    time = soup.find("span",{"class":"duration"}).text
    
    
    container['title']=title
    container['year']=year
    container['level']=m_num
    container['time']=time
    container['movie_type']=m_type
    container['speciality']=spec
    result.append(container)

    #netflix = pd.DataFrame({"title":title,"year":year,"level":m_num, "time":time, "movie_type":m_type, "speciality":spec},index=[0])

print(result)
```
