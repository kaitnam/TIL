```python
import time   
import requests
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm

result_4 = []   # 전체 크롤링 데이터를 담을 그릇
# comments = ""

# 수집할 글 갯수 정하기
# number = 5  

# 수집한 url 돌면서 데이터 수집
for u in tqdm(url_list_1):
    driver = webdriver.Chrome('C:/Users/katenam/chromedriver.exe')
    url = (u)
    driver.get(url)
    time.sleep(2)
    
    # 크롤링

#     try : 
    # iframe 접근
    driver.switch_to.frame('mainFrame')

    target_info = {}  # 개별 블로그 내용을 담을 딕셔너리 생성

    # 제목 크롤링 시작
    try:
        overlays = ".se-module.se-module-text.se-title-text"                                 
        tit = driver.find_element_by_css_selector(overlays)          # title
        title = tit.text
    except:
        title = ' '


    # 글쓴이 크롤링 시작
    try:
        overlays = ".link.pcol2"                               
        nick = driver.find_element_by_css_selector(overlays)         # nickname
        nickname = nick.text
    except:
        nickname = ' '

    # 날짜 크롤링
    try:
        overlays = ".se_publishDate.pcol2"                                 
        date = driver.find_element_by_css_selector(overlays)         # datetime
        datetime = date.text
    except:
        datetime = ' '


    # 내용 크롤링
    try:
        
        overlays = ".se-component.se-text.se-l-default"              # one content                  
        contents = driver.find_elements_by_css_selector(overlays)    # contents

        content_list = []
        for content in contents:
            content_list.append(content.text)

        content_str = ' '.join(content_list)                         # content_str
    except:
        content_str = ' '

    # 댓글 크롤링
    try: 
        a = u[-12:]
        comment = driver.find_element_by_css_selector('a#Comi'+a)
        comment.click()
        time.sleep(1)

        overlays = ".u_cbox_contents"                          
        rep = driver.find_elements_by_css_selector(overlays)

        comments = ""
        for i in rep:

            comments = comments + i.text #댓글이 여러개 있기 때문에 for문을 돌려서 댓글을 누적해야함. 그래서 comments + i.text를 한거임.

    except:
        comments = ' '
        
    # 공감수
    try:
        b = u[-12:]
        like = driver.find_element_by_css_selector('#Sympathy'+b+ '> div > span > em.u_cnt._count')  
  
        likes = like.text
    except:
        likes = ' '

        
        
    # 글 하나는 target_info라는 딕셔너리에 담기게 되고,
    target_info['url'] = url
    target_info['title'] = title
    target_info['nickname'] = nickname
    target_info['datetime'] = datetime
    target_info['content'] = content_str
    target_info['comments'] = comments
    target_info['like'] = likes

    result_4.append(target_info)
    print(target_info)

    driver.close()
```
