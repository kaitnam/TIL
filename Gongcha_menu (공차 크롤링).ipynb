```python
from bs4 import BeautifulSoup
from selenium import webdriver
from pprint import pprint
import re

driver=webdriver.Chrome('C:/Users/kaitnam/Documents/PYTHON/chromedriver.exe')
numbers=[]
for i in range(1001,1013,1):
    url=('http://www.gong-cha.co.kr/brand/menu/product.php?c=00'+str(i))
    driver.get(url)
    html=driver.page_source

    soup=BeautifulSoup(html, 'html.parser')
    wrap=soup.findAll('div',{"class":"pro_list_wrap"})

    for li in wrap:
        lii=li.findAll('li')
        for a in lii:
            atag=a.find('a')
            idx=atag['n']
            numbers.append(idx)
pprint(numbers)
    
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from pprint import pprint
import zlib
import requests

result=[]
for number in numbers:
    container=dict()
    driver=webdriver.Chrome('C:/Users/kaitnam/Documents/PYTHON/chromedriver.exe')
    url=('http://www.gong-cha.co.kr/brand/menu/productView.php?no={}'.format(number))
    session = requests.Session()
    req = session.get(url)
    req.encoding = 'utf-8'
    driver.close()

    soup=BeautifulSoup(req.text, 'html.parser')

    info=soup.find('div', {"class":"info"})
    name=info.find('p', {"class":"tit"}).text
    tr=info.findAll('tr')[2]
    volume=str(tr.findAll('td')[1].text)+'g'
    kcal=str(tr.findAll('td')[2].text)+'kcal'
    sugars=str(tr.findAll('td')[3].text)+'g'
    protein=str(tr.findAll('td')[4].text)+'g'
    sat_FAT=str(tr.findAll('td')[5].text)+'g'
    sodium=str(tr.findAll('td')[6].text)+'mg'
    caffeine=str(tr.findAll('td')[7].text)+'mg'

    container['name']=name
    container['volume']=volume
    container['kcal']=kcal
    container['sugars']=sugars
    container['protein']=protein
    container['sat_FAT']=sat_FAT            
    container['sodium']=sodium
    container['caffeine']=caffeine
    result.append(container)

pprint(result)
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from pprint import pprint
import zlib
import requests

url=('http://www.gong-cha.co.kr/brand/menu/productView.php?no=198')
session = requests.Session()
req = session.get(url)
req.encoding = 'utf-8'

result=[]
container=dict()
soup=BeautifulSoup(req.text, 'html.parser')

info=soup.find('div', {"class":"info"})
name=info.find('p', {"class":"tit"}).text
tr=info.findAll('tr')[2]
volume=str(tr.findAll('td')[1].text)+'g'
kcal=str(tr.findAll('td')[2].text)+'kcal'
sugars=str(tr.findAll('td')[3].text)+'g'
protein=str(tr.findAll('td')[4].text)+'g'
sat_FAT=str(tr.findAll('td')[5].text)+'g'
sodium=str(tr.findAll('td')[6].text)+'mg'
caffeine=str(tr.findAll('td')[7].text)+'mg'

container['name']=name
container['volume']=volume
container['kcal']=kcal
container['sugars']=sugars
container['protein']=protein
container['sat_FAT']=sat_FAT            
container['sodium']=sodium
container['caffeine']=caffeine
result.append(container)

pprint(result)
```
