```python
review_1 = review.dropna() # 결측치 삭제
review_1
```

```python
review_1.isnull().sum()
```

```python
review_2 = review_1.drop_duplicates() # 같은 열 제거(내용이 같은 열이 몇개 있었음)
review_2
```

```python
review_total=[]

for i in review_2['Comments']:
    review_total.append(i)

review_total
```

```python
result = ' '.join(s for s in review_total) # review_total 리스트를 문자열로 형태소 분석을 하기 위해 문자열로 바꾸기
print(result)
```

```python
# Okt 이용
```

```python
from collections import Counter
review_counts = Counter(result)
```

```python
review_counts.most_common(50)
```

```python
from konlpy.tag import Okt


okt = Okt()
nouns = okt.nouns(result) # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
```

```python
c
```

```python
c.most_common(50)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(font_path='malgun', width=800, height=800, scale=2.0, max_font_size=250, background_color="white")
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```

```python
wc.to_file('Socar.png')
```

```python
# Verb 추출
```

```python
okt = Okt()
verb = okt.pos(result, norm=True, stem=True) # 모든 형태소 추출
words = [x for x, y in verb if y in ['Verb']] # 추출된 값 중 동사만 추출

c = Counter(words)
 
#Okt = Text(myList_filter, name="Okt")
c.most_common(50)
```
