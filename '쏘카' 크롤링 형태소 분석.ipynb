```python
# 형태소 분석(morphological analysis)은 자연어의 최소 의미인 형태소를 식별하는 언어 처리 과정. 
# 구문 분석(syntax analysis)은 주어, 동사, 목적어등의 문장성분을 판별해 문장성분에 따른 문장 구조를 분석하는 과정.

# <종류>
# Hannanum (한나눔)
# Kkma (꼬꼬마)
# Komoran (코모란)
# Mecab (메카브)
# Open Korean Text (Okt)
```

```python
# nouns : 명사 추출
# morphs : 형태소 추출
# pos : 품사 부착
```

```python
from konlpy.tag import *

hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
okt = Okt()
```

```python
import pandas as pd

review = pd.read_csv("C:/Users/katenam/Socar_Review_Final.csv")
review.head()
```

```python
review.isnull().sum()
```

```python
review_total=[]

for i in review['Comments']:
    review_total.append(i)

review_total
```

```python
result = ' '.join(s for s in review_total) # review_total 리스트를 문자열로 형태소 분석을 하기 위해 문자열로 바꾸기
print(result)
```

### 한나눔

```python
from collections import Counter

nouns = hannanum.nouns(result)
words = [n for n in nouns if len(n) > 1]
c = Counter(words)
c.most_common(50)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(font_path='malgun', width=800, height=800, scale=2.0, max_font_size=250, background_color="white")
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```

```python
wc.to_file('Socar_hannanum_noun.png')
```
